```python
import pandas as pd
import numpy as np
```

```python
# 결과 불러오기
df1_path = 'C:/Users/manci/Desktop/멀티잇_부트캠프/Mini/최종결과_files/review_preprocessing_final_excel.xlsx'

df = pd.read_excel(df1_path)
```

```python
df.head()
```

```python
# 감성 분석 실행
from transformers import pipeline

pipe = pipeline("text-classification", model="WhitePeak/bert-base-cased-Korean-sentiment")
```

```python
# 예시 적용
sample = list(df.iloc[0:5, 1])
print(sample)
print('-'*80)

'''
LABEL_0: negative
LABEL_1: positive
'''

result = pipe(sample)
print(result)
```

```python
### 한국어 분석 모델(huggingface)
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 토그나이저 생성
tokenizer = AutoTokenizer.from_pretrained("WhitePeak/bert-base-cased-Korean-sentiment")

# 분류 모델 생성
model = AutoModelForSequenceClassification.from_pretrained("WhitePeak/bert-base-cased-Korean-sentiment")

# pipline 모델 생성
classifier = pipeline('text-classification', model = model, tokenizer = tokenizer)

# 샘플 생성
test = list(df.iloc[0:5, 1])

# 감성 분석 실행
result = classifier(test)
print(result)
```

```python
# 샘폴 데이터 생성 --> list 자료형으로 생성
sample = list(df.iloc[0:5, 1])
print(sample)
print('-'*80)

# 감성 분석 실행
result = classifier(sample)
print(result)
print('-'*80)

# 결과 확인하기
for data in result:
    label = data['label']
    if label=='LABEL_1':
        print('긍정')
    else:
        print('부정')
```

```python
# 전체 리뷰 데이터
text_data = list(df.loc[:, '리뷰'])

# 감성 분석 실행(파인튜닝)
result = classifier(text_data)
print(result)
```

```python
# LABEL_1은 '긍정', LABEL_0은 '부정'으로 매핑
label_mapping = {
    'LABEL_1': '긍정',
    'LABEL_0': '부정'
}

# 결과 값을 label 칼럼에 추가
new_labels = []  # 새로운 라벨을 저장할 리스트 생성
for data in result:
    # 결과 값의 label을 가져와서 label_mapping에서 매핑된 값을 찾음
    new_label = label_mapping[data['label']]
    # 새로운 라벨(긍정, 부정)을 new_labels에 추가
    new_labels.append(new_label)

# 새로운 라벨 리스트를 데이터프레임의 label 칼럼에 할당
df['label'] = new_labels
```

```python
import pandas as pd

# CSV 파일 경로
file_path = 'C:/Users/manci/Desktop/멀티잇_부트캠프/Mini/최종결과_files/Mini_huggingface_final_찐막_csv.csv'
output_file_path = 'C:/Users/manci/Desktop/멀티잇_부트캠프/Mini/최종결과_files/Mini_호감도_csv.csv'

# CSV 파일 읽기
df = pd.read_csv(file_path)

# 매장명 별로 긍정과 부정 레이블의 갯수를 세는 함수
def count_positive_negative_by_store(df):
    store_counts = {}

    # 각 행을 반복하며 매장명과 레이블을 확인
    for index, row in df.iterrows():
        store_name = row['매장명']
        label = row['label']

        if store_name not in store_counts:
            store_counts[store_name] = {'긍정': 0, '부정': 0}

        if label == '긍정':
            store_counts[store_name]['긍정'] += 1
        elif label == '부정':
            store_counts[store_name]['부정'] += 1

    return store_counts

# 호감도를 계산하는 함수
def calculate_favorability(positive_count, negative_count):
    if positive_count + negative_count == 0:
        return 0
    favorability = positive_count / (positive_count + negative_count)
    return favorability

# 각 매장별로 호감도를 계산하고 출력하는 함수
def calculate_and_add_store_favorability(df):
    store_counts = count_positive_negative_by_store(df)
    favorability_list = []

    for index, row in df.iterrows():
        store_name = row['매장명']
        counts = store_counts[store_name]
        positive_count = counts['긍정']
        negative_count = counts['부정']
        favorability = calculate_favorability(positive_count, negative_count)
        favorability_list.append(favorability)

    df['호감도'] = favorability_list
    return df

# 호감도 계산 및 추가
df = calculate_and_add_store_favorability(df)

# 수정된 데이터프레임을 새로운 CSV 파일로 저장
df.to_csv(output_file_path, index=False)

# 업데이트된 데이터프레임 출력
print(df[['매장명', '리뷰', '호감도']])
```

```python
import pandas as pd

# CSV 파일 경로
file_path = 'C:/Users/manci/Desktop/멀티잇_부트캠프/Mini/최종결과_files/Mini_호감도_csv.csv'
output_file_path ='C:/Users/manci/Desktop/멀티잇_부트캠프/Mini/최종결과_files/Mini_최종결과_csv.csv'

df = pd.read_csv(file_path)

df1 = df[df.호감도 != 1]
```

```python
df1.매장명.unique()
```

```python
# 호감도에 가중치를 적용하는 함수
def apply_weight_to_sentiment(df1):
    weights = df1['가중치']
    weighted_sentiment = df1['호감도'] * weights
    df1['가중치_호감도'] = weighted_sentiment

    return df1

# 호감도에 가중치를 적용하여 데이터프레임 업데이트
df1 = apply_weight_to_sentiment(df1)

# 수정된 데이터프레임을 새로운 CSV 파일로 저장
df1.to_csv(output_file_path, index=False)

# 업데이트된 데이터프레임 출력
print(df1[['매장명', '리뷰', '호감도', '가중치_호감도']])
df1.sort_values(by='호감도', ascending=False)
```

```python
# 매장명별로 가중치_호감도를 추출
idx = df1.groupby('매장명')['가중치_호감도'].idxmax()
unique_store_df1 = df1.loc[idx]

# 가중치_호감도의 합 계산 (분모)
total_weighted_sentiment = unique_store_df1['가중치_호감도'].sum()

# 각 매장의 가중치_호감도를 백분율로 계산
unique_store_df1['백분율'] = (unique_store_df1['가중치_호감도'] / total_weighted_sentiment) * 100

# 상위 매장 출력 (가중치 호감도 순으로 정렬)
sorted_unique_store_df1 = unique_store_df1.sort_values(by='가중치_호감도', ascending=False)

print("가중치_호감도가 가장 높은 매장 (백분율 포함):")
for index, row in sorted_unique_store_df1.iterrows():
    print(f"매장명: {row['매장명']}, 호감도: {(row['호감도']*100):.2f}%, 가중치_호감도: {(row['가중치_호감도']):.2f}")
```

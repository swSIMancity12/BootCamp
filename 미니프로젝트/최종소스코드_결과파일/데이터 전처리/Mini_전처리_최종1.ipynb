```python
### ì˜ì–´ ë¦¬ë·° ì°¾ê¸°
import re
'''
# ì˜ì–´ë¡œ ì‹œì‘í•˜ëŠ” ë¬¸ì¥ì„ ì°¾ëŠ” í•¨ìˆ˜ ì •ì˜
def starts_with_english(text):
    if re.match(r'^[A-Za-z]', text):
        return True
    return False

english_start_reviews = df[df['reviews'].apply(starts_with_english)]
english_start_reviews
'''
# ì˜ì–´ ë¬¸ì¥ì¸ì§€ í™•ì¸í•˜ëŠ” ì •ê·œ í‘œí˜„ì‹ íŒ¨í„´
english_pattern = re.compile(r'^[A-Za-z]')

# ë¦¬ë·°ì˜ ì²« ë¬¸ì¥ì´ ì˜ì–´ë¡œ ì‹œì‘í•˜ëŠ”ì§€ í™•ì¸í•˜ëŠ” í•¨ìˆ˜
def is_english_first_sentence(review):
    if pd.isna(review):
        return False
    # ì²« ë¬¸ì¥ ì¶”ì¶œ
    first_sentence = re.split(r'[.!?]', review.strip())[0]
    return bool(english_pattern.match(first_sentence.strip()))

# ì˜ì–´ë¡œ ì‹œì‘í•˜ëŠ” ì²« ë¬¸ì¥ë§Œ ì¶”ì¶œ
english_reviews_df = df[df['reviews'].apply(is_english_first_sentence)]
english_reviews_df
```

```python
### pip install pandas googletrans==4.0.0-rc1 ì„¤ì¹˜ í›„ ì‚¬ìš©
# ì˜ì–´ ë¦¬ë·° --> í•œê¸€ë¡œ ë³€ê²½

from googletrans import Translator

# Translator ê°ì²´ ìƒì„±
translator = Translator()

# ì˜ì–´ ë¦¬ë·°ë¥¼ í•œê¸€ë¡œ ë²ˆì—­í•˜ëŠ” í•¨ìˆ˜
def translate_to_korean(text):
    try:
        translated = translator.translate(text, src='en', dest='ko')
        return translated.text
    except Exception as e:
        print(f"Error translating text: {text}\n{e}")
        return text  # ë²ˆì—­ì— ì‹¤íŒ¨í•œ ê²½ìš° ì›ë˜ í…ìŠ¤íŠ¸ ë°˜í™˜

# ë²ˆì—­ ì ìš©
english_reviews_df['reviews'] = english_reviews_df['reviews'].apply(translate_to_korean)
english_reviews_df['reviews']
```

```python
# 1702ë²ˆì§¸ ì¹¼ëŸ¼ ìŠ¤í˜ì¸ì–´ --> í•œê¸€ë¡œ ë³€ê²½
def translate_to_korean_if_spanish(text):
    try:
        # ë¦¬ë·°ì˜ ì–¸ì–´ ê°ì§€
        detected_lang = translator.detect(text).lang
        if detected_lang == 'es':  
            translated = translator.translate(text, src='es', dest='ko')
            return translated.text
        else:
            return text 
    except Exception as e:
        print(f"Error translating text: {text}\n{e}")
        return text

# ìŠ¤í˜ì¸ì–´ ë¦¬ë·° --> í•œêµ­ì–´ë¡œ ë²ˆì—­
english_reviews_df['reviews'] = english_reviews_df['reviews'].apply(translate_to_korean_if_spanish)
english_reviews_df['reviews']
```

```python
### "reviews" ì¹¼ëŸ¼ì— íŠ¹ì • ë‹¨ì–´ ë³€í™˜ í•¨ìˆ˜ ë§Œë“¤ê¸°

# ì¢‹ì•˜ì–´ìš”, ë‚˜ì˜ì§€ ì•Šì•„ìš”, ì¶”ì²œí•˜ì§€ ì•Šì•„ìš”, ë§›ìˆì–´ìš”, ì¹œì² í•´ìš”, ê°„ë‹¨í•˜ê²Œ ë§¥ì£¼í•œì” í–ˆì–´ìš”, í‰ê°€í•˜ì§€ ì•Šê² ìŠµë‹ˆë‹¤ ë¡œ ë³€í™˜í•  ë‹¨ì–´ ëª©ë¡
good_words = ['êµ¿', 'êµ¿êµ¿', 'êµ³', 'êµ³êµ³', 'ã„±ã„±', 'êµ¬ì›ƒ', 'ìµì˜¤',
              'êµ¿ê¸‹', 'ê·¯', 'êµ¿~', 'ì¡°ì•¼', 'êµ¿ğŸ˜„', 'ã…', '^~^', 'ã„±ã„±ã„±',
              'Good', 'Gooooooood', 'Gooooooood!!!!!', 'ì¢‹ì¢‹', 'ì¢‹ì€', 'ì¡°ì•„ìš”',]
notbad_words = ['ë‚«ë°·', 'ë¬´ë‚œ', 'ë…¸ë§', 'ì˜¤', 'Not bad']
bad_words = ['ë¹„ì¶”']
delicious_words = ['ì•¼ë¯¸', 'ë§›ì§‘', 'ë§›', 'ğŸ»ğŸ—', 'ì¡´ë§›', 'ğŸ—', 
                   'JMT', 'Fuggin Deli Shush ğŸ™ğŸŒâ¤ï¸', 'ì•¼ë¯¸~']
kind_words = ['ì¹œì ˆ']
simple_words = ['ê°„ë§¥']

def replace_reviews(text):
    if text in good_words:
        return 'ì¢‹ì•˜ì–´ìš”'
    elif text in notbad_words:
        return 'ë‚˜ì˜ì§€ ì•Šì•„ìš”'
    elif text in bad_words:
        return 'ì¶”ì²œí•˜ì§€ ì•Šì•„ìš”'
    elif text in delicious_words:
        return 'ë§›ìˆì–´ìš”'
    elif text in kind_words:
        return 'ì¹œì ˆí•´ìš”'
    elif text in simple_words:
        return 'ê°„ë‹¨í•˜ê²Œ ë§¥ì£¼í•œì” í–ˆì–´ìš”'
    else:
        text
    return text

# one_two_dfì˜ reviews ì¹¼ëŸ¼ì— ëŒ€í•´ í•¨ìˆ˜ ì ìš©
df['reviews'] = df['reviews'].apply(replace_reviews)

# ì›ë˜ dfì— ë³€í™˜í•œ ë°ì´í„° ë°˜ì˜
#df.update(one_two_df)
```

```python
# íŒ¨í„´ë§Œ ì§€ìš°ëŠ” ì „ì²˜ë¦¬ í•¨ìˆ˜
def preprocessing_review_non_korea(review):
    review = review.replace('\n', '')
    # í•œê¸€ë§Œ
    cleaned_review = re.sub(r'[^ê°€-í£]', ' ', review)
    # ê³µë°± ì—¬ëŸ¬ê°œ --> í•˜ë‚˜ì˜ ê³µë°±ìœ¼ë¡œ ì¶•ì†Œ
    cleaned_review = re.sub(r'\s+', ' ', cleaned_review).strip()
    return cleaned_review
```

```python
# preprocessing_review_non_korea --> ì˜ˆì‹œ ì ìš©
sample = df.iloc[0, 1]
text = preprocessing_review_non_korea(sample)
print(sample)
print('-'*80)
print(text)
```

```python
# íŒ¨í„´ë§Œ ì§€ìš°ëŠ” ì „ì²˜ë¦¬ í•¨ìˆ˜ ì‹¤í–‰
cleaned_reviews_1 = []

for review in tqdm(df.loc[:, 'reviews']):
    cleaned_text = preprocessing_review_non_korea(review)
    cleaned_reviews_1.append(cleaned_text)

print(cleaned_reviews_1)
```
